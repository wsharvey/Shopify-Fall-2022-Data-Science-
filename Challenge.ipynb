Question 1a)

As shown above, there seems to be an issue with shop_id = 42 and shop_id = 78, and their order amount number. In order to fix this issue, I would propose removing the rows that correspond to shop_id = 42. It seems as if every data entry for shop_id = 42 is the same, so these rows wouldn't be helpful for our analysis. The only different value is the order_id, and every other value is the same. As far as shop_id = 78, it is possible that the order_amount was mistakenly entered in cents, so that can be converted by dividing each order_amount with shop_id = 78 by 100. This is done below.

```python
converted = data[data['shop_id'] == 78]
converted['order_amount'] = converted['order_amount'] / 100
clean_data = data[(data['shop_id'] != 42) & (data['shop_id'] != 78)]
clean = pd.concat([clean_data, converted])
```

Question 1b) and 1c)

Lets take a look at the different metrics we can use, now that we've cleaned the data.

```python
clean['order_amount'].describe()
```

Mean doesn't seem like the best choice, as it seems to be much higher than the 50th percentile, and is influenced heavily by the outliers. In this case, I would instead prefer to report median as the best metric because it isn't affected by the large outliers. This report would lead to a value of 284, which is much more realistic than the previous report.
